```python
import pandas as pd
```

```python
my_list = [5, 3, 5, 1, 0, 7, 8, 1, 9, 5, 5, 7, 3, 4, 8, 9, 5, 3, 3, 2, 6, 1]
```

```python
my_series = pd.Series(my_list)
```

```python
my_series
```

```python
my_sample = my_series.sample(n=5)
my_sample
```

```python
sample_with_replacement = my_series.sample(n=10, replace=True)
sample_with_replacement
```

```python
sample_1 = my_series.sample(n=5,
                            replace= True,
                            random_state=42)
sample_1
```

```python
sample_2 = my_series.sample(n=5,
                            replace= True,
                            random_state=42)
sample_2
```
